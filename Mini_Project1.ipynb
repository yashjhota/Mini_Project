````python
#Connect Google Drive to Colab to store data:
from google.colab import drive
drive.mount('/content/drive')
````

````python
import requests
import pandas as pd

url = "https://api.github.com/repos/apache/hadoop/commits"
response = requests.get(url)
data = response.json()

# Extract required fields
commit_data = []
for commit in data:
    commit_data.append([
        commit['sha'],
        commit['commit']['author']['name'],
        commit['commit']['author']['date'],
        commit['commit']['message']
    ])

# Create a DataFrame
df = pd.DataFrame(commit_data, columns=["SHA", "Author", "Date", "Message"])
df.head()
````

````python
df.to_csv("/content/drive/MyDrive/github_commits.csv", index=False)
print("Data saved to Google Drive!")
````

````python
df = pd.read_csv("/content/drive/MyDrive/github_commits.csv")
print(df.head())
````

````python
import sqlite3

# Create in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Load data into SQL table
df.to_sql('github_commits', conn, if_exists='replace', index=False)
````

ðŸ“Œ **Query 1**: Count Total Records

````python
query = "SELECT COUNT(*) FROM github_commits"
result = pd.read_sql(query, conn)
result
````

ðŸ“Œ **Query 2**: Find Most Active Authors

````python
query = """
SELECT Author, COUNT(*) as CommitCount
FROM github_commits
GROUP BY Author
ORDER BY CommitCount DESC
LIMIT 5
"""
result = pd.read_sql(query, conn)
result
````

ðŸ“Œ **Query 3**: Search for Specific Words in Commit Messages

````python
query = """
SELECT Message FROM github_commits
WHERE Message LIKE '%bug%'
"""
result = pd.read_sql(query, conn)
result
````

ðŸ“Œ **Query 4**: Get the latest 10 commits

````python
query = """
SELECT * FROM github_commits
ORDER BY Date DESC
LIMIT 10
"""
result = pd.read_sql(query, conn)
result
````

ðŸ“Œ **Query 5**:Count the number of commits per day

````python
query = """
SELECT Date, COUNT(*) AS CommitCount
FROM github_commits
GROUP BY Date
ORDER BY Date ASC
LIMIT 3
"""
result = pd.read_sql(query, conn)
result
````

ðŸ“Œ **Query 6**: Find the most frequent words in commit messages

````python
query = """
SELECT Message, COUNT(*) as Frequency
FROM github_commits
GROUP BY Message
ORDER BY Frequency DESC
LIMIT 10
"""
result = pd.read_sql(query, conn)
result
````

ðŸ“Œ **Query 7**:Count unique authors contributing to the project

````python
query = """
SELECT COUNT(DISTINCT Author) AS UniqueAuthors
FROM github_commits
"""
result = pd.read_sql(query, conn)
result
````

# ðŸ“Š **Data Visualizations**

ðŸ“Š Plot commit activity over time:

````python
import matplotlib.pyplot as plt

df['Date'] = pd.to_datetime(df['Date'])
df['Date'].dt.date.value_counts().sort_index().plot(kind='bar', figsize=(12, 5))

plt.xlabel('Date')
plt.ylabel('Number of Commits')
plt.title('GitHub Commits Over Time')
plt.xticks(rotation=45)
plt.show()
````

ðŸ“Š Top 5 authors contributing most commits:

````python
df['Author'].value_counts().head(5).plot(kind='bar', color='skyblue')

plt.xlabel('Author')
plt.ylabel('Commit Count')
plt.title('Top 5 Authors with Most Commits')
plt.show()
````

ðŸ“Š Commit Activity Over Time (Line Plot):

````python
df['Date'] = pd.to_datetime(df['Date'])
commits_per_day = df.groupby(df['Date'].dt.date).size()

plt.figure(figsize=(12, 5))
plt.plot(commits_per_day.index, commits_per_day.values, marker='o', linestyle='-')

plt.xlabel('Date')
plt.ylabel('Number of Commits')
plt.title('Daily Commit Activity Over Time')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

ðŸ“Š  Most Frequent Words in Commit Messages (Word Cloud)

````python
from wordcloud import WordCloud

text = ' '.join(df['Message'].dropna())
wordcloud = WordCloud(width=400, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Commit Messages")
plt.show()
````

ðŸ“Š  Number of Commits by Author (Pie Chart):

````python
top_authors = df['Author'].value_counts().head(5)
plt.pie(top_authors, labels=top_authors.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Top 5 Authors by Number of Commits')
plt.show()
````

ðŸ“Š  Most Active Authors (Bar Chart):

````python
df['Author'].value_counts().head(10).plot(kind='barh', color='coral', figsize=(10, 5))
plt.xlabel('Number of Commits')
plt.ylabel('Author')
plt.title('Top 10 Most Active Authors')
plt.gca().invert_yaxis()  # To show highest at the top
plt.show()
````
